```python
import corr
import cv2
import matplotlib.pyplot as plt
from skimage.feature import blob_log
```

```python
frames = corr.load_example_experiment()
test_frame = frames[0]
```

```python
plt.imshow(test_frame, cmap="gray")
```

```python
blobs_log = blob_log(test_frame, min_sigma=5, max_sigma=20, num_sigma=5, threshold=0.04)
```

```python
fig, ax = plt.subplots()
ax.imshow(test_frame, cmap="gray")
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r, linewidth=1, fill=False, color="red")
    ax.add_patch(c)
plt.show()
```

```python
img8 = corr.float32_to_uint8(test_frame)
denoised = cv2.bilateralFilter(img8, d = 10, sigmaColor=20, sigmaSpace=50)
blobs_log = blob_log(denoised, min_sigma=5, max_sigma=20, num_sigma=5, threshold=0.04)
fig, ax = plt.subplots()
ax.imshow(denoised, cmap="gray")
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r, linewidth=1, fill=False, color="red")
    ax.add_patch(c)
plt.show()
```
